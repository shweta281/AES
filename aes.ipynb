```python
def gf_multiply(a, b, mod=0x11B):
    """Multiply two numbers in GF(2^8) using the modulo polynomial x^8 + x^4 + x^3 + x + 1 (0x11B)."""
    result = 0
    while b:
        if b & 1:
            result ^= a
        a <<= 1
        if a & 0x100:
            a ^= mod
        b >>= 1
    return result

def gf_inverse(byte):
    """Find the multiplicative inverse in GF(2^8) using the extended Euclidean algorithm."""
    if byte == 0:
        return 0  # 0 has no inverse in GF(2^8), but it's mapped to itself in AES
    a, b, x0, x1 = byte, 0x11B, 1, 0
    while a != 1:
        quotient = b // a
        b, a = a, b ^ gf_multiply(quotient, a)
        x0, x1 = x1, x0 ^ gf_multiply(quotient, x1)
    return x0

def affine_transformation(byte):
    """Apply the affine transformation used in AES S-Box and inverse S-Box."""
    affine_matrix = 0xF8  # 11111000 in binary
    result = 0
    for i in range(8):
        result ^= ((byte >> i) & 1) * ((affine_matrix >> i) & 1)
    return result ^ 0x05  # XOR with the fixed vector for inverse S-Box

def generate_inv_s_box():
    """Generate the AES Inverse S-Box dynamically."""
    inv_s_box = [0] * 256
    for i in range(256):
        inv_s_box[i] = affine_transformation(gf_inverse(i))
    return inv_s_box

# Generate and print the Inverse S-Box
inv_s_box = generate_inv_s_box()
print(inv_s_box)
```
